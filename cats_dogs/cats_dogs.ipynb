## Define a smaller Model

````python
class DogsAndCatsClassifierSmall(nn.Module):
    def __init__(self, img_size):
        super().__init__()
        
        self.conv1 = nn.Conv2d(3, 32, 3) # input channels, output, kernel size
        self.conv2 = nn.Conv2d(32, 64, 3)
        self.conv3 = nn.Conv2d(64, 128, 3)
        
        self.pool = nn.MaxPool2d(2, 2)
        
        # find out the size of the following fc layer
        x = torch.randn(3, img_size, img_size).view(-1, 3, img_size, img_size)
        self._to_linear = None
        self.convs(x)
        self.fc1 = nn.Linear(self._to_linear, 128)
        self.fc2 = nn.Linear(128, 2) # 2 classes
        
    def convs(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))
  
        if self._to_linear == None:
            self._to_linear = x[0].shape[0] * x[0].shape[1] * x[0].shape[2] # x[0] first element of batch

        return x
        
    def forward(self, x):
        
        x = self.convs(x)
        x = x.view(-1, self._to_linear) # flatten x
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        
        return x
````

## Define the training Loop

````python
def train(model, epoch, device, train_loader, criterion, optimizer):
    
    train_loss = []
    predictions = []
    correct = 0
    
    model.train()
    # loop over batches
    loop = tqdm(enumerate(train_loader), leave=False, total=len(train_loader))
    for idx, (img, label) in loop:
            
        # set to device
        img, label = img.to(device), label.to(device)
        
        optimizer.zero_grad()
        
        # forward pass
        y_hat = model(img)
    
        # loss
        loss = criterion(y_hat, label)
        train_loss.append(loss)
        
        # backward
        loss.backward()
        
        # update the weights
        optimizer.step()
    
        y_pred = y_hat.argmax(dim=-1, keepdims=True)
        predictions.append(y_pred)
        correct += y_pred.eq(label.view_as(y_pred)).sum().item()
        train_acc = torch.sum(y_pred == label.view_as(y_pred))
        accuracy = train_acc/batch_size
        
        # update progess bar
        loop.set_description(f'Train Epoch {epoch+1}/{epochs}')
        loop.set_postfix(loss=loss.item(), acc=accuracy.item())
    
    # total training loss over all batches
    train_loss = torch.mean(torch.tensor(train_loss))
    epoch_accuracy = correct/len(train_loader.dataset)
    
    loop.set_postfix(train_loss=train_loss.item(), accuracy=100. * epoch_accuracy)
    return train_loss, epoch_accuracy
````

````python
def validation(model, device, val_loader, criterion):
    
    val_loss = []
    predictions = []
    correct = 0
    
    model.eval()
    # loop over batches
    with torch.no_grad():
        loop = tqdm(enumerate(val_loader), leave=False, total=len(val_loader))
        for idx, (img, label) in loop:
            
            # set to device
            img, label = img.to(device), label.to(device)
        
            # forward pass
            y_hat = model(img)
    
            # loss
            loss = criterion(y_hat, label)
            val_loss.append(loss)
    
            y_pred = y_hat.argmax(dim=-1, keepdims=True)
            predictions.append(y_pred)
            correct += y_pred.eq(label.view_as(y_pred)).sum().item()
            val_acc = torch.sum(y_pred == label.view_as(y_pred))
            accuracy = val_acc/batch_size
    
            # update progess bar
            loop.set_description(f'Val Epoch {epoch+1}/{epochs}')
            loop.set_postfix(loss=loss.item(), acc=accuracy.item()*100)
    
    # total training loss over all batches
    val_loss = torch.mean(torch.tensor(val_loss))
    epoch_accuracy = correct/len(val_loader.dataset)
    loop.set_postfix(val_loss=val_loss.item(), accuracy=100. * epoch_accuracy)
    
    return val_loss, epoch_accuracy, predictions
````

# Train the Model

## Set the Hyperparameters

````python
# hyperparameters
batch_size = 64
epochs = 5
learning_rate = 0.001
````

## Load the Data

````python
train_loader = torch.utils.data.DataLoader(train_data,
                   shuffle=True, batch_size=batch_size)
test_loader = torch.utils.data.DataLoader(test_data,
                   shuffle=False, batch_size=batch_size)
````

````python
print(f'image dataloader item: {next(iter(train_loader))[0].shape}')
print(f'label dataloader length of item: {len(next(iter(train_loader))[1])}')
````

````python
# Visualize one training batch
for images, labels in train_loader:
    fig, ax = plt.subplots(figsize = (10, 10))
    ax.set_xticks([])
    ax.set_yticks([])
    ax.imshow(make_grid(images, 4).permute(1,2,0))
    break
````

````python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)

model = DogsAndCatsClassifier.to(device)
#model = DogsAndCatsClassifierSmall(img_size).to(device)

# define the optimizer
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

# define loss 
criterion = nn.CrossEntropyLoss()

train_losses_epoch = []
test_losses_epoch = []
train_acc_epoch = []
test_acc_epoch = []
preds_epoch = []
min_loss = np.inf

start_time = time.time()
for epoch in range(epochs):
    
    train_loss, train_epoch_accuracy = train(model, epoch, device, train_loader, criterion, optimizer)
    train_losses_epoch.append(train_loss)
    train_acc_epoch.append(train_epoch_accuracy)
    
    test_loss, test_epoch_accuracy, predictions = validation(model, device, test_loader, criterion)
    test_losses_epoch.append(test_loss)
    test_acc_epoch.append(test_epoch_accuracy)
    preds_epoch.append(predictions)
    
    # save model if loss is decreasing
    if train_loss < min_loss:
        min_loss = train_loss
        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'loss': test_loss,
            }, f'{model.__class__.__name__}_epoch={epoch:02d}_acc={train_epoch_accuracy:.3f}.pt')
        print("model saved")
    
    print(f'epoch {epoch+1} finished')
    print(f"epoch {epoch+1}/{epochs}, train loss={train_loss}, test loss={test_loss}")
    print(f"epoch {epoch+1}/{epochs}, train acc={train_epoch_accuracy}, test acc={test_epoch_accuracy}")

print(f"Training took: {(time.time()-start_time):.3f}s")
````
